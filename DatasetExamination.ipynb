```python
import pandas as pd
import numpy as np

np.set_printoptions(threshold=np.inf)
pd.set_option('display.max_rows', None)  # Display all rows
pd.set_option('display.max_columns', None)  # Display all columns

# Load the dataset
dataset = pd.read_csv('data_train.csv')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 16), dpi=160)
dataset.hist()

plt.show()
```

```python
plt.figure(figsize=(20, 20))
for i, col in enumerate(dataset.columns):
    plt.subplot(4,5,i+1)
    plt.hist(dataset[col], alpha=0.3, color='r', density=True)
    plt.title(col)
    plt.xticks(rotation='vertical')
```

```python
dataset.plot(kind='box', subplots=True, layout=(6,5), sharex=False, sharey=False)
plt.show()
```

```python
correlation = dataset.corr()
fig = plt.figure(figsize=(10,10), dpi=80)
ax = fig.add_subplot(111)
cax = ax.matshow(correlation, vmin=-1, vmax=1, cmap=plt.cm.PuBu)
fig.colorbar(cax)

# Set the number of ticks to match the number of columns in the dataset
ticks = np.arange(0, len(dataset.columns), 1)

ax.set_xticks(ticks)
ax.set_yticks(ticks)

# Set the tick labels to the column names of the dataset
ax.set_xticklabels(dataset.columns, rotation=90)
ax.set_yticklabels(dataset.columns)

plt.show()
```
